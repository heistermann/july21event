```python
import os
import wradlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.spatial import KDTree
#import helpers as h
import datetime as dt
import matplotlib.dates as mdates
import matplotlib as mpl
from matplotlib.colors import BoundaryNorm, LinearSegmentedColormap, to_hex
from matplotlib.lines import Line2D
```

```python
%matplotlib inline
```

```python
rwdirpattern = "rw/raa01-rw_10000-%s-dwd---bin.gz"
```

```python
proj = wradlib.georef.epsg_to_osr(32632) # UTM32N
```

```python
# Get coordinates
rwcoords = wradlib.georef.get_radolan_grid(900,900)
x_rw, y_rw = wradlib.georef.reproject(rwcoords[:,:,0],
                                      rwcoords[:,:,1],
                                      projection_source=wradlib.georef.create_osr("dwd-radolan"),
                                      projection_target=proj)
```

```python
dataset, inLayer = wradlib.io.open_vector("shapes/europe_countries.shp")
countries, keys = wradlib.georef.get_vector_coordinates(inLayer, dest_srs=proj)

dataset, inLayer = wradlib.io.open_vector("shapes/bundeslaender_wgs84.shp")
laender, keys = wradlib.georef.get_vector_coordinates(inLayer, dest_srs=proj)
#bbox = list(inLayer.GetExtent())
#bbox[0] = bbox[0]-3000
```

```python
def proc_rw(dtimes, rwdirpattern):
    rwprec = np.zeros((len(dtimes), 900, 900)) * np.nan
    for i, dtime in enumerate(dtimes):
        fpath = rwdirpattern % dtime.strftime("%y%m%d%H%M")
        try:
            data, meta = wradlib.io.read_radolan_composite(fpath, missing=np.nan)
            rwprec[i] = data
            print(".", end="")
        except:
            raise()
            print("Could not read: %s" % fpath)
    print("")
    return rwprec
```

```python
tstart = "2021-07-13 00:50:00"
tend   = "2021-07-16 00:50:00"
```

```python
dtimes = wradlib.util.from_to(tstart, tend, 3600)
```

```python
rwprec = proc_rw(dtimes, rwdirpattern)
```

```python
plt.rc('font', **{'size'   : 14})
fig, ax = plt.subplots(figsize=(12,12), nrows=1, ncols=1)

ax.set_aspect("equal")

gridcolor = "lightgray"
linecolor = "white"
#fontsize = 18

xoffset = 100000
yoffset = 5400000

cmap = plt.cm.nipy_spectral#plt.cm.magma

bounds = np.arange(0, 210, 20)
lenmap = len(bounds) - 1
colornums = [int(i*(cmap.N-1)/(lenmap-1)) for i in (np.arange(lenmap))]
colors = [cmap(i) for i in colornums]
listedcmap = mpl.colors.ListedColormap(colors)
norm = BoundaryNorm(bounds, listedcmap.N)


pm = plt.pcolormesh(x_rw, y_rw, np.ma.masked_invalid(rwprec.sum(axis=0)), cmap=listedcmap, norm=norm)
plt.colorbar(pm, shrink=0.75)
wradlib.vis.add_lines(ax, countries, color=linecolor, lw=2.5)
wradlib.vis.add_lines(ax, laender, color=linecolor, lw=1)

plt.grid()

plt.text(xoffset+10000, yoffset + 480000, "%s - %s" % (tstart, tend), fontsize=20, color="tab:red",
        bbox=dict(facecolor='white', edgecolor='none', alpha=0.8))



plt.ticklabel_format(axis="x", useOffset=xoffset)
plt.ticklabel_format(axis="y", useOffset=yoffset)

plt.xlim(xoffset, xoffset+500000)
plt.ylim(yoffset, yoffset+500000)
plt.xlabel("UTM Easting (m)")
plt.ylabel("UTM Northing (m)")

plt.title("Event depth, RW product")

plt.savefig("figs/july21event.png", dpi=300)
```
